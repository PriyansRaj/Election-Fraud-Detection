```python
from sklearn.impute import KNNImputer
```

```python
imputer = KNNImputer(n_neighbors=10)
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns
```

```python
df[num_cols] = imputer.fit_transform(df[num_cols])
```

```python
for col in cat_cols:
    df[col].fillna(df[col].mode()[0],inplace =True)
```

```python
winners = df.loc[df.groupby(['st_name','year'])['totvotpoll'].idxmax(),['st_name','year',"pc_name","cand_name","partyname",'totvotpoll']].reset_index(drop=True)
```

```python
winners = winners.sort_values(['year','st_name']).reset_index(drop=True)
winners
```

```python
candidates_counts = df.loc[df.groupby(['year','totvotpoll'])['electors'].idxmax(),["st_name",'year','totvotpoll','pc_name','electors']].reset_index(drop=True)
```

```python
candidates_counts = candidates_counts.sort_values(['year','totvotpoll']).reset_index(drop=True)
candidates_counts
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['voter_turnout'] = sum(df['totvotpoll']/df['electors']) *100
```

```python
df['vote_share'] = df.groupby(['year','pc_name'])['totvotpoll'].transform(lambda x: (x/x.sum())*100)
```

```python

party_votes = df.groupby(['year', 'st_name', 'partyname'])['vote_share'].mean().reset_index()


def calc_pdi(group):
    shares = group['vote_share'] / 100
    return (shares ** 2).sum()

pdi = party_votes.groupby(['year', 'st_name']).apply(calc_pdi).reset_index(name='pdi')

df = df.merge(pdi, on=['year', 'st_name'], how='left')
```

```python

df = df.drop(columns=['pdi_y'],axis=1)
```

```python
df =df.rename({'pdi_x':'pdi'})
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder,OrdinalEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
```

```python
low_card_cat = ['st_name', 'pc_type', 'cand_sex']
high_card_cat = ['partyname', 'partyabbre']
numeric_cols = ['year', 'pc_no', 'totvotpoll', 'electors', 'voter_turnout', 'vote_share', 'pdi_x']

# Drop unnecessary columns
df = df.drop(columns=['pc_name', 'cand_name'])
```

```python
for col in high_card_cat:
    freq_map = df[col].value_counts().to_dict()
    df[col+"_freq"] = df[col].map(freq_map)
numeric_cols += [col + '_freq' for col in high_card_cat]
```

```python
cat_transformer = Pipeline(steps=[
    ('imputer',SimpleImputer(strategy="most_frequent")),
    ('encoder',OneHotEncoder(handle_unknown='ignore'))
])
num_transformer = Pipeline(steps=[
     ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])
```

```python
preprocessor = ColumnTransformer(
    transformers = [
        ('cat',cat_transformer,low_card_cat),
        ('num',num_transformer,numeric_cols)
    ]
)
```

```python
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor)
])
```

```python
X_preprocessed = pipeline.fit_transform(df)
```
